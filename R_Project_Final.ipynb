# **Data Splitting**

```r
# Splitting Data

data_sample <- sample(2, nrow(data), replace = TRUE, prob = c(0.8, 0.2))
```

```r
train = data[data_sample == 1,]
test = data[data_sample == 2,]
```

```r
# Head of training data

head(train)
```

```r
# Dimension of training data

dim(train)
```

```r
# Head of training data

head(test)
```

```r
# Removing target variable from test dataset

test_features <- test[ ,-c(8)]
test_target= test[ ,c(8)]
head(test_features)
```

# **Model Fitting**
## **Logistic Regression**

```r
log_regression <- glm(formula = AQI ~ ., data=train, family = binomial())
```

```r
summary(log_regression)
```

```r
# Area Under the Curve Plot

library(pROC)

lr_predict <- predict(log_regression, test_features)
auc_gbm = roc(test_target, lr_predict, plot = TRUE)
```

From above graph, we can conclude that our **logistic regression model has an accuracy of about 67%.**

## **Decision Tree**

```r
library(party)

decision_tree <- ctree(formula = AQI ~ ., data = train)
```

```r
plot(decision_tree)
```

```r
print(decision_tree)
```

```r
decision_predict <- predict(decision_tree, test_features)
auc_gbm = roc(test_target, decision_predict, plot = TRUE)
```

From above graph, we can conclude that our **decision tree model has an accuracy of about 87%**.

# **Random Forest**

```r
library(randomForest)

forest <- randomForest(formula = AQI ~ ., data = train)
```

```r
print(forest)
print(importance(forest, type = 2))
```

```r
forest_predict <- predict(forest, test_features)
auc_gbm = roc(test_target, forest_predict, plot = TRUE)
```

From above, we can conclude that **Random Forest works the best with over 95% of accuracy.**
